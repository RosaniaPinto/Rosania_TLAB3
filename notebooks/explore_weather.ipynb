```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather_data = pd.concat([weather1, weather2], ignore_index=True)

# Save to CSV
weather_data.to_csv("../data/weather/weather_data.csv", index=False)
```

```python
# TODO: view first 5 rows of your data
weather_data.head()
```

```python
# TODO: describe your numeric columns
weather_data.describe()
```

```python
# TODO: display the null values in your dataframe
weather_data.isnull()
```

```python
# TODO: display the null values in your dataframe
print(weather_data.isnull().sum())
```

```python
# TODO: BEGIN YOUR EDA!

#Lineplot of Rain Max Across The Years
weather_data.plot(x='year', y='rain_max', kind='line')
```

```python
#Lineplot 
#Wind Across The Years For Wind Maximum And Wind Average
weather_data.plot(x='year', y='wind_max', kind='line')
weather_data.plot(x='year', y='wind_avg', kind='line')
```

```python
#Lineplot 
#Humidity Across The Years For Humidity Maximum And Humidity Average
weather_data.plot(x='year', y='hum_max', kind='line')
weather_data.plot(x='year', y='hum_min', kind='line')
```

```python
#Lineplot 
#Temperature Across The Years For Temperature Maximum, Temperature Average, And Temperature Minimum 
weather_data.plot(x='year', y='temp_max', kind='line')
weather_data.plot(x='year', y='temp_avg', kind='line')
weather_data.plot(x='year', y='temp_min', kind='line')
```
