```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather_data = pd.concat([weather1, weather2], ignore_index=True)

# Save to CSV
weather_data.to_csv("../data/weather/weather_data.csv", index=False)
```

```python
# TODO: view first 5 rows of your data
weather_data.head()
```

```python
# TODO: describe your numeric columns
weather_data.describe()
```

```python
# TODO: display the null values in your dataframe
weather_data.isnull()
```

```python
# TODO: display the null values in your dataframe
print(weather_data.isnull().sum())
```

```python
# TODO: BEGIN YOUR EDA!

# Histogram of coffee production
# Set the plot size
plt.figure(figsize=(10, 6))

# Create the histogram
# Add bins or KDE if desired
sns.histplot(weather_data, bins=20, kde=False)  

# Add labels and title
plt.title('Distribution of Weather')
plt.xlabel('Total Production')
plt.ylabel('Frequency')

# Rotate ticks for readability
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
sns.histplot(weather_data, kde=False)  
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(weather_data)
plt.title('Rain, temperature, and humidity')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(weather_data)
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
